# Relationship between Shots Taken and Goal Scored

## This section of the study aims to determine if there is a correlation between the number of shots a team takes during a game, and if it corresponds to the team scoring more goals on average

```python
#create table to compare the correlation between shots taken and the result
shots_win_ratio = pd.DataFrame({'Home Goals': df['FTHG'], 'Home Shots': df['HS'], 'Away Goals': df['FTAG'], 'Away Shots': df['AS']})
shots_win_ratio
```

```python
#correlation of home goals and home shots
shots_win_ratio.plot.scatter(x = 'Home Shots', y = 'Home Goals')
plt.title('Home Team Shots vs Home Team Goals')
```

```python
#Correlation Table for home goals to home shots
home_goals_ratio = pd.DataFrame({'Home Goals': df['FTHG'], 'Home Shots': df['HS']})
sns.heatmap(home_goals_ratio.corr(), annot = True)

plt.title('Correlation of Home Team Shots & Home Team Goals')
plt.show()
```

```python
#correlation of away goals and away shots
shots_win_ratio.plot.scatter(x = 'Away Shots', y = 'Away Goals')
plt.title('Away Team Shots vs Away Team Goals')
```

```python
#Correlation Table for away goals to away shots
away_goals_ratio = pd.DataFrame({'Away Goals': df['FTAG'], 'Away Shots': df['AS']})
sns.heatmap(away_goals_ratio.corr(), annot = True)

plt.title('Correlation of Away Team Shots & Away Team Goals')
plt.show()
```

From the two figures above, it can be seen that there is a very small correlation between number of shots a team takes and the number of goals that, that team would score. This is further highlighted by the correlation figures as the home teams shots that correlate to the home team scoring is only 0.33, whilst only being 0.25 for the away teams shots that correlates to their goals.

Thus, on average despite home teams having a higher chance of scoring as opposed to their away team counterparts, the average correlation of only 0.29 for both teams means allows us to conclude that having a high number of shots doesn't greatly improve a teams chance of scoring.

Up to this point, this study however does not consider the types of shots that the teams are taking nor does it look that the quality of chances that are being created by each team, which could contribute to the low values of correlation

# Attack vs Defence

```python
#create table to store number of goals teams have scored and conceded
table = pd.DataFrame()
table['Teams'] = df['HomeTeam'].unique()

table = table.sort_values(by = 'Teams', ascending = True)
table
```

```python
home_stats = df.groupby('HomeTeam').sum()
home_stats = home_stats.reset_index()
home_stats.rename(columns={'HomeTeam': 'Teams'}, inplace = True)
home_stats
```

```python
away_stats = df.groupby('AwayTeam').sum()
away_stats = away_stats.reset_index()
away_stats.rename(columns={'AwayTeam': 'Teams'}, inplace = True)
away_stats
```

```python
#collecting the number of goals scored and conceded for each team when they played at home
table = table.merge(home_stats, how = 'inner', on = 'Teams')

#drop unnecessary columns
table.drop(columns = ["HTHG", "HTAG", "HS" ,"AS", "HST", "AST", "HF", "AF", "HC", "AC", "HY", "AY", "HR", "AR"], inplace = True)
table
```

```python
table = table.merge(away_stats, how = 'inner', on = 'Teams')

#drop unnecessary columns
table.drop(columns = ["HTHG", "HTAG", "HS" ,"AS", "HST", "AST", "HF", "AF", "HC", "AC", "HY", "AY", "HR", "AR"], inplace = True)
table
```

```python
table.rename(columns = {'FTHG_x': 'Goals_scored_at_Home', 'FTAG_x': 'Goals_conceded_at_Home', 'FTHG_y': 'Goals_conceded_Away', 'FTAG_y': 'Goals_scored_Away'}, inplace = True)
table['Goals For'] = table['Goals_scored_at_Home'] + table['Goals_scored_Away']
table['Goals Against'] = table['Goals_conceded_at_Home'] + table['Goals_conceded_Away']
table['Goal Difference'] = table['Goals For'] - table['Goals Against']
```

```python
table
```

```python
#Clubs position based on attack
attack_table = table.sort_values(by = 'Goals For', ascending = False)
attack_table
```

```python
attack_table['Goals For'].plot(kind = 'bar', stacked = True,   xlabel = 'Teams', ylabel = 'Goals Scored')
plt.title('Goals Scored by Teams')
```

```python
#Clubs position based on defence
defence_table = table.sort_values(by = 'Goals Against', ascending = True)
defence_table
```

```python
defence_table['Goals Against'].plot(kind = 'bar', stacked = True,  xlabel = 'Teams', ylabel = 'Goals Conceded')
plt.title('Goals Conceded by Teams')
```

More often than not, the data seems to illustrate that teams are more likely to perform better in the league if they have a better attack than a better defence. This is because the best teams in the leagues e.g. teams like Man City, or Arsenal are able to counter potential defence weaknesses by simply outscoring their opponents.  Moreover, fom the results of the season, the teams that on had more goals scored, even if they conceded a lot of goals, still tended to finish above those teams that would not score as many goals but were relatively good defensivelyWhile having a good defence does play an important factor in how a team performs, most teams tend to benefit more from being 'top heavy' (having a stronger attack than defence')
